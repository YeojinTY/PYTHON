```python
import random

n = random.randrange(1, 4)  # 1부터 3까지의 값 생성
if n == 1:
    computer_choice = 'left'
elif n == 2:
    computer_choice = 'center'
else:
    computer_choice = 'right'

user_choice = str(input('어디를 공격하시겠어요??? (왼,중,오): '))  # 사용자 입력 받기

# 사용자 입력을 영어로 변환하여 비교
if user_choice == '왼':
    user_choice = 'left'
elif user_choice == '중':
    user_choice = 'center'
elif user_choice == '오':
    user_choice = 'right'
else:
    print("잘못된 입력입니다. '왼', '중', '오' 중 하나를 선택하세요.")
    exit()

if user_choice == computer_choice:
    print('fail')  # 컴퓨터가 막음
else:
    print('congratulation!')  # 성공

print('컴퓨터의 수비 위치:', computer_choice)
```
